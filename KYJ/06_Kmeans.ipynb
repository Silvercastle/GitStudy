### K-Means : 비지도학습, 군집화(Clustering)

- 사전 준비
  1.  데이터 클러스터링의 주제 결정
  2. 클러스터 개수 결정
  3. 데이터 준비
  4. 클러스터링 수행

- 수행 과정
  - 중심에 가까운 데이터를 클러스터에 포함
  - 중심을 클러스터의 중앙으로 이동
  - 위 과정 반복

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
from sklearn.cluster import KMeans
```

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(columns=['x', 'y'])
```

```python
df = pd.DataFrame(np.random.randint(0,20,size=(30, 2)), columns=list("xy"))
```

```python
df.head(30)
```

```python
sb.lmplot(x='x', y='y', data=df, fit_reg=False, scatter_kws={"s":100})
plt.title('K-means Example')
plt.xlabel('x')
plt.ylabel('y')
```

```python
points = df.values
kmeans = KMeans(n_clusters=4).fit(points)
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
df['cluster'] = kmeans.labels_
df.head(30)
```

```python
sb.lmplot(x='x', y='y', data=df, fit_reg=False, scatter_kws={"s": 150}, hue="cluster")
```
