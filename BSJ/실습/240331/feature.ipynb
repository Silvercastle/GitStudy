```python
import tensorflow as tf
import matplotlib.pyplot as plt

# TFRecord 파일 경로
tfrecord_path = r'Z:\5_AI\HDAT\BSJ\실습\240331\~\tensorflow_datasets\cats_vs_dogs\4.0.0\cats_vs_dogs-train.tfrecord-00000-of-00008'

# TFRecordDataset을 사용하여 TFRecord 파일 읽기
dataset = tf.data.TFRecordDataset(tfrecord_path)

# for item in dataset:
#     print(item)

# TFRecord 파일 내의 데이터 파싱하기 (예: 이미지, 레이블 등)
# 이 파싱 방법은 데이터가 어떻게 저장되어 있는지에 따라 달라집니다.
# 이 예제는 데이터가 이미지와 레이블로 구성되어 있다고 가정합니다.
def parse_record(record):
    # TFRecord 파일에서 읽은 데이터를 파싱합니다.
    features = {
        'image': tf.io.FixedLenFeature([], tf.string),
        'label': tf.io.FixedLenFeature([], tf.int64),
    }
    parsed_record = tf.io.parse_single_example(record, features)
    
    # 이미지를 디코딩합니다. (이 예제는 JPEG 이미지를 가정합니다.)
    image = tf.image.decode_jpeg(parsed_record['image'], channels=3)
    
    # 레이블을 반환합니다.
    label = parsed_record['label']
    
    return image, label

# 데이터셋에 파싱 함수를 적용합니다.
dataset = dataset.map(parse_record)

# 데이터셋을 순회하여 데이터 확인
for image, label in dataset:
    # 이미지와 레이블 처리 또는 사용
    # 여기에서는 이미지와 레이블을 출력합니다.
    print("Label:", label.numpy())
    # 이미지를 확인하려면 matplotlib 등의 라이브러리를 사용할 수 있습니다.
    # plt.imshow(image.numpy())
```
