### 수강신청 준비하기

```python
import pandas as pd

df = pd.read_csv('data/enrolment_1.csv')

# 여기에 코드를 작성하세요
df['status']='allowed'
# condition1,2
df.loc[(df['year']==1) & (df['course name']=='information technology'),'status'] = 'not allowed'
df.loc[(df['year']==4) & (df['course name']=='commerce'),'status'] = 'not allowed'

df['boolean'] = 0
df.loc[df['status']=='not allowed', 'boolean'] = 0
df.loc[df['status']=='allowed', 'boolean'] = 1
df2 = df.groupby('course name').sum(numeric_only=True)
df2 = df2[['boolean']]
df2 = df2.reset_index()
course_list = []
for i in range(0,len(df2.index)):
    if df2['boolean'][i] < 5:
        course_list.append(df2['course name'][i])
for j in range(0,len(df.index)):
    if df['course name'][j] in course_list:
        df['status'][j] = 'not allowed'
df = df.drop(columns='boolean')

# 테스트 코드
df
```

```python
# 모범답안
import pandas as pd

df = pd.read_csv('data/enrolment_1.csv')
df["status"] = "allowed"

# 조건 1
boolean1 = df["course name"] == "information technology"
boolean2 = df["year"] == 1
df.loc[boolean1 & boolean2, "status"] = "not allowed"

# 조건 2
boolean3= df["course name"] == "commerce"
boolean4= df["year"] == 4
df.loc[boolean3 & boolean4, "status"] = "not allowed"

# 조건 3
allowed = df["status"] == "allowed"
course_counts = df.loc[allowed, "course name"].value_counts()
closed_courses = list(course_counts[course_counts < 5].index)
for course in closed_courses:
    df.loc[df["course name"] == course, "status"] = "not allowed"

# 테스트 코드
df
```

### 강의실 배정하기 I

```python
import pandas as pd

df = pd.read_csv('data/enrolment_2.csv')

# 여기에 코드를 작성하세요
allowed = df['status'] == 'allowed'
course_counts = df.loc[allowed, 'course name'].value_counts()
course_counts

# 조건 1
auditorium_list = list(course_counts[course_counts >= 80].index)

# 조건 2
large_room_list = list(course_counts[(course_counts >= 40) & (course_counts < 80)].index)

# 조건 3
medium_room_list = list(course_counts[(course_counts >= 15) & (course_counts < 40)].index)

# 조건 4
small_room_list = list(course_counts[(course_counts >= 5) & (course_counts < 15)].index)

for i in range(0,len(df.index)):
    if df['course name'][i] in auditorium_list :
        df.loc[i, 'room assignment'] = 'Auditorium'
    elif df['course name'][i] in large_room_list :
        df.loc[i, 'room assignment'] = 'Large room'
    elif df['course name'][i] in medium_room_list :
        df.loc[i, 'room assignment'] = 'Medium room'
    elif df['course name'][i] in small_room_list :
        df.loc[i, 'room assignment'] = 'Small room'

# 조건 5
df.loc[df['status']=='not allowed', 'room assignment'] = 'not assigned'

# 테스트 코드
df

```
