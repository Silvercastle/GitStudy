### concat

```python
card = pd.concat([card_2017,card_2018])
card.info()
```

```python
card['연월'].unique()
```

```python
card.describe(include='all')
```

```python
object_columns = card.columns[card.dtypes=='object']

for i in range(0,len(object_columns)):
    print(card[object_columns[i]].unique())
```

### 이용금액과 이용자수가 많은 업종과 지역 Top 10

```python
# 이용자수 Top10
df1 = card.groupby('업종명').sum(numeric_only=True).sort_values('이용자수', ascending=False).head(10)
df1 = df1.reset_index()
df1
```

```python
plt.rc('font', family='Malgun Gothic')
df1.plot(kind='bar', x='업종명', y='이용자수')
plt.title('업종명별 이용자수 Top10')
```

```python
# 이용금액 Top10
df2 = card.groupby('업종명').sum(numeric_only=True).sort_values('이용금액', ascending=False).head(10)
df2 = df2.reset_index()
df2
```

```python
plt.rc('font', family='Malgun Gothic')
df2.plot(kind='bar', x='업종명', y='이용금액')
plt.title('업종명별 이용금액')
```

### 인당 이용금액이 많은 지역 Top10

```python
card['지역'] = card['시군구명'] + card['읍면동명']
card.head()
```

```python
df3 = card.groupby('지역').sum(numeric_only=True)
df3['인당 이용금액'] = df3['이용금액'] / df3['이용자수']
df3 = df3.sort_values('인당 이용금액', ascending=False).head(10)
df3 = df3.reset_index()
df3
```

```python
plt.rc('font', family = 'Malgun Gothic')
df3.plot(kind='bar', x='지역', y='인당 이용금액')
```

```python
card.head()
```

### 각 지역에서 어떤 업종이 활성화 되었는지

```python
df4 = card.pivot_table(values='이용자수',
                       index='지역',
                       columns='업종명',
                       aggfunc='sum')
df4.head()
```

```python
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(15,10))
# df4.plot(kind='bar')
sns.heatmap(df4, cmap='BuPu', linewidths=0.5)
```

## 카페 이용자수와 유동인구의 관계

### population 전처리

```python
# population 전처리
# card와 merge를 위해 데이터 변환
# 1. 성별
population['성별'] = population['성별'].replace('남', '남성')
population['성별'] = population['성별'].replace('여', '여성')
```

```python
# 2. 연월일
population['연월일'] = population['연월일'].astype(str)
population['연월일'] = pd.to_datetime(population['연월일']).astype(str)
# -- 칼럼명 통일
population.columns = ['연월', '시군구명', '읍면동명', '성별', '연령대', '방문인구']
```

```python
card.head()
```

```python
population.head()
```

```python
population.info()
```

```python
# join
df5 = pd.merge(left=card, right=population, how='left', on=['연월','시군구명','읍면동명','성별'])
df5
```

```python
df5['업종명'].unique()
```

```python
df5.describe(include='all')
```

### 카페 이용자수와 유동인구의 관계

```python
df6 = df5[df5['업종명']=='비알콜 음료점업']
df6 = df6.reset_index()
df6 = df6.drop('index', axis=1)
df6
```

```python
df6 = df6.groupby('지역').sum(numeric_only=True).sort_values('방문인구', ascending=False).head(10)
df6 = df6.reset_index()
df6
```

```python
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(14,8))
df6.plot(kind='bar', x='지역', y=['이용자수', '방문인구'])
```

```python
df7 = df5[df5['업종명']=='비알콜 음료점업']
df7 = df7.reset_index()
df7 = df7.drop('index', axis=1)
df7 = df7.groupby('지역').sum(numeric_only=True).sort_values('방문인구', ascending=False)
df7 = df7.reset_index()
```

```python
df7
```

```python
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(14,8))
plt.scatter(df7['지역'], df7['이용자수'], label='이용자수')
plt.scatter(df7['지역'], df7['방문인구'], label='방문인구')
plt.legend()
plt.grid(True)
plt.title('카페 이용자수와 방문인구의 관계')
plt.xticks(rotation=90)
```
