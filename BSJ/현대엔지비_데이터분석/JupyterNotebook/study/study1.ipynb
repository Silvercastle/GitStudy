```python
import pandas as pd
```

```python
file_path = './data/졸업생의 진로 현황(전체).xlsx'

data = pd.read_excel(file_path)
```

```python
data.columns
```

```python
use_cols = ['지역', '정보공시 \n 학교코드', '학교명', '졸업자.2', '(특수목적고)과학고 진학자.2', '(특수목적고)외고ㆍ국제고 진학자.2']

data2 = data[use_cols]
data2
```

```python
data2.columns = ['reg', 'code', 'name', 'grads', 'sci', 'intl']
```

```python
data3 = data2.drop( 0 )
```

```python
data3.isna().sum()
```

```python
data3[ data3['reg'].isna() ]
```

```python
data3.loc[588, 'reg'] = '서울특별시 마포구'
data3.loc[3011, 'reg'] = '부산광역시 해운대구'
```

```python
data3.loc[ 588 ]
```

```python
data4 = data3.dropna().copy()
```

```python
data4.dtypes
```

```python
data4.loc[1, 'sci']
```

```python
data4['grads'] = pd.to_numeric(data4['grads'])
data4['sci'] = pd.to_numeric(data4['sci'])
data4['intl'] = pd.to_numeric(data4['intl'])
```

```python
data4.dtypes
```

```python
data4['spc_N'] = data4['sci'] + data4['intl']
data4['spc_R'] = (data4['spc_N'] / data4['grads'] * 100).round(2)
```

```python
cond1 = data4['grads'] >= 100
data4[cond1].sort_values(by='spc_R', ascending=False)[:20]
```

```python
# apply
# 한 열에 모든 데이터들에게 똑같은 코드를 일괄적으로 적용

def get_sido( x ):
    return x.split( ' ' )[0]

def get_sigungu( x ):
    return x.split( ' ' )[1]

data4['sido'] = data4['reg'].apply(get_sido)
data4['sigungu'] = data4['reg'].apply(get_sigungu)
```

```python
# 서울특별시 내에서 진학률이 높은 20개
temp = data4['sido'] == '서울특별시'
data4[ temp ].sort_values(by='spc_R', ascending=False)[:20]
```

```python
data4['reg'].unique()
```

```python
# 집계
# 여러 데이터를 하나의 숫자나 문자로 표현
# mean, sum, max, min, count

data4['spc_R'].mean()
data4['grads'].mean()
```

```python
# group by
# 특정 카테고리 별로 묶어서 집계
# 결과가 1차원

data4.groupby( 'sido' )['spc_R'].mean()
temp = data4.groupby( 'sido' )['grads'].sum()
temp.sort_values(ascending=False)
```

```python
# 학교 개수
temp = data4.groupby( 'sido' )['name'].count()
temp.sort_values(ascending=False)
```
